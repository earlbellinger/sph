```python
import numpy as np
import scipy.special as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import animation
import cartopy
import cartopy.crs as ccrs
```

```python
from easing import easing
```

```python
l=5; m=5
lon = np.linspace(0, 2*np.pi, 200) - np.pi
lat = np.linspace(-np.pi/2, np.pi/2, 500)
colat = lat + np.pi/2
d = np.zeros((len(lon), len(colat)), dtype=np.complex64)
for j, yy in enumerate(colat):
    for i, xx in enumerate(lon):
        d[i, j] = sp.sph_harm(m, l, xx, yy)
drm = np.transpose(np.real(d)) #only interested in real components
```

```python
fig = plt.figure(figsize=(10, 10),tight_layout = {'pad': 0})
#inclination = 60 
plotcrs = ccrs.Mercator() #PlateCarree() #ccrs.Orthographic(0, 90 - inclination)
ax = plt.subplot(projection=plotcrs)

vlim = np.max(np.abs(drm))
ax.pcolormesh(lon*180/np.pi, lat*180/np.pi, 
              drm, transform=ccrs.PlateCarree(),
              cmap='PuOr', vmin=-vlim*3, vmax=vlim*3);

#ax.autoscale_view()
#plt.savefig('l4m3.png',dpi=300)
```

```python
nframes = 32 
seconds = 2. 
interval = seconds/nframes

def init():
    return 

def animate(i):
    drm = np.transpose(np.real(d*np.exp(-1.j*(2.*np.pi*float(i) / float(nframes)))))
    ax.clear()
    ax.pcolormesh(lon*180/np.pi, lat*180/np.pi, drm,
         transform=ccrs.PlateCarree(), cmap='PuOr', vmin=-vlim, vmax=vlim)
    ax.autoscale_view()
    return

#Animate it and save!
fig = plt.figure(figsize=(2, 2), tight_layout = {'pad': 0})
ax = plt.subplot(projection=plotcrs)
anim = animation.FuncAnimation(fig, animate, init_func=init, 
                               frames=nframes, interval=interval, 
                               blit=False)
anim.save('l4m3.gif', dpi=200, fps=1/interval, writer='imagemagick')
```
